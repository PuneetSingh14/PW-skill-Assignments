Q3. Explain the following terms:
1. The difference in np.array, np.asarray and np.asanyarray.
2. The difference between Deep copy and shallow copy.

- np.array: This is the base function to create a NumPy array. It takes in a list or other iterable and creates a new array object. If the input is already an array, it will create a copy.

- np.asarray: This function is similar to np.array, but it will not create a copy if the input is already an array. It will return the original array if it's already an array.

- np.asanyarray: This function is similar to np.asarray, but it will also allow non-NumPy arrays (like lists or tuples) as input and will not create a copy if the input is already an array.

2. Deep copy and shallow copy:

- Shallow copy: A shallow copy creates a new object and then (to the extent possible) inserts references into it to the objects found in the original. This means that if you modify a sub-object in the copy, it will also be modified in the original.

- Deep copy: A deep copy creates a new object and then, recursively, inserts copies into it of the objects found in the original. This means that if you modify a sub-object in the copy, it will not be modified in the original.

Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
# Create a time-series dataset in a Pandas DataFrame
date_range = pd.date_range('2024-01-01', periods=100)
temperature = np.random.uniform(20, 30, 100)
humidity = np.random.uniform(50, 90, 100)
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})
df
```

```python
# Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(df['Date'], df['Temperature'], color='red', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', marker='o')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Temperature and Humidity Over Time')
plt.grid(True)
plt.show()
```

Q17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'

```python
# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Create a scatter plot with Seaborn
plt.figure(figsize=(8, 8))
sns.scatterplot(x=x, y=y, hue=np.where((x >= 0) & (y >= 0), 'Quadrant I',
                                        np.where((x < 0) & (y >= 0), 'Quadrant II',
                                       np.where((x < 0) & (y < 0), 'Quadrant III', 'Quadrant IV'))),
                palette=['red', 'green', 'blue', 'yellow'])

# Add legend, label axes, and set title
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

Q18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.sampledata.iris import flowers
from bokeh.plotting import figure, output_file, show

# Generate x values
x = np.linspace(0, 2*np.pi, 100)

# Calculate corresponding y values (sine wave)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function",
           x_axis_label='x',
           y_axis_label='sin(x)')

# Add a line renderer to the figure
p.line(x, y, legend_label="sin(x)", line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3
p.grid.grid_line_color = "navy"

# Show the results
show(p)
```
